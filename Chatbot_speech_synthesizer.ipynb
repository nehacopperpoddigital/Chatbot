```python
import speech_recognition as sr
import pyttsx3

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Initialize the speech recognizer
r = sr.Recognizer()

# Defining a function to recognize speech and return text
def recognize_speech():
    with sr.Microphone() as source:
        print("Speak now")
        audio = r.listen(source)

    try:
        text = r.recognize_google(audio)
        return text
    except sr.UnknownValueError:
        return "Sorry, I couldn't understand you"
    except sr.RequestError as e:
        return f"Sorry, there was an error with the speech recognition service: {e}"

# Defining a function to synthesize speech from text
def synthesize_speech(text):
    engine.say(text)
    engine.runAndWait()

# Get input from the user
user_input = recognize_speech()
print(f"User said: {user_input}")

# Generate a response
if "hello" in user_input:
    response = "Hi there!"
elif "how are you" in user_input:
    response = "I'm doing well, thank you. How about you?"
elif "what is your name" in user_input:
    response = "I am Tony."
elif "what is your favourite color" in user_input:
    response = "I like blue, what is your favourite color?"
elif "whould you like to chat with me" in user_input:
    response = "Yes"
elif "would you like to play with me" in user_input:
    response = "Yes, which game do you want to play?"
else:
    response = "Sorry, I didn't understand what you said."

# Synthesize the response and speak it
synthesize_speech(response)
print(f"Bot said: {response}")
```

```python
import speech_recognition as sr
import pyttsx3
import nltk
nltk.download()
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Initialize the speech recognizer
r = sr.Recognizer()

# Define a function to recognize speech and return text
def recognize_speech():
    with sr.Microphone() as source:
        print("Speak now")
        audio = r.listen(source)

    try:
        text = r.recognize_google(audio)
        return text
    except sr.UnknownValueError:
        return "Sorry, I couldn't understand you"
    except sr.RequestError as e:
        return f"Sorry, there was an error with the speech recognition service: {e}"

# Define a function to synthesize speech from text
def synthesize_speech(text):
    engine.say(text)
    engine.runAndWait()

# Define a function to generate a response
def generate_response(input_text):
    # Tokenize the input text
    tokens = word_tokenize(input_text.lower())

    # Define a list of response keywords and their corresponding responses
    response_keywords = {
        "hello": ["Hi there!", "Hello!", "Howdy!"],
        "how are you": ["I'm doing well, thanks. How about you?", "I'm great. How are you doing?", "I'm fine. How are you?"],
        "goodbye": ["Goodbye!", "See you later!", "Farewell!"],
        "thank you": ["You're welcome!", "No problem.", "My pleasure."]
    }

    # Check if any of the response keywords appear in the input text
    for keyword in response_keywords:
        if wordnet.synsets(keyword):
            if keyword in tokens:
                return response_keywords[keyword][0]

    # If no response was generated, return a default response
    return "I'm sorry, I don't understand what you're asking."

# Get input from the user
user_input = recognize_speech()
print(f"User said: {user_input}")

# Generate a response
response = generate_response(user_input)

# Synthesize the response and speak it
synthesize_speech(response)
print(f"Bot said: {response}")
```
